```python
import pandas as pd
from datetime import datetime

print("🚀 Iniciando extracción masiva de datos (Fuente: Global Power Plant DB)...")

# Esta es una fuente de datos abiertos masiva y muy estable
url_global = "https://raw.githubusercontent.com/wri/global-power-plant-database/master/output_database/global_power_plant_database.csv"

try:
    # 1. Lectura directa (Usamos low_memory=False para archivos grandes)
    df_raw = pd.read_csv(url_global, low_memory=False)
    
    # 2. Filtrado y Construcción de tu formato
    # Tomamos 500 plantas de energía
    df_datos = df_raw.head(500).copy()
    
    df_final = pd.DataFrame({
        "fuente_sitio": "World Resources Institute",
        "dataset": "global_power_plant_database",
        "url_origen": "https://purl.org/wri/grid/gpv2",
        "grupo": "energia-y-sustentabilidad-1",
        "tema": "Impacto ambiental de la energia",
        "fecha_extraccion": datetime.now().strftime("%Y-%m-%d %H:%M:%S"),
        "pais": df_datos['country_long'],
        "region": df_datos['country'],
        "periodo": 2026, # Año actual de la auditoría
        "indicador": "capacidad_generacion",
        "categoria_energia": df_datos['primary_fuel'].apply(lambda x: "Renovable" if x in ['Solar', 'Wind', 'Hydro', 'Biomass', 'Geothermal'] else "Fosil"),
        "tecnologia": df_datos['primary_fuel'],
        "actor": df_datos['owner'].fillna("General Utility"),
        "item": df_datos['name'].astype(str) + " - " + df_datos['primary_fuel'].astype(str),
        "valor": df_datos['capacity_mw'].astype(float),
        "unidad": "MW"
    })

    # 3. Guardado local para tu GitHub
    nombre_archivo = "datos_auditoria_global.csv"
    df_final.to_csv(nombre_archivo, index=False, encoding='utf-8')
    
    print(f"✅ ¡Por fin! Se extrajeron {len(df_final)} registros exitosamente.")
    print(f"📂 Archivo listo para subir: {nombre_archivo}")
    print("\n--- Vista Previa del Resultado ---")
    print(df_final[['pais', 'tecnologia', 'valor', 'unidad']].head(10))

except Exception as e:
    print(f"❌ Error inesperado: {e}")
```
