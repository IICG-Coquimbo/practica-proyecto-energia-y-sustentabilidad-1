```python
# 1 # Apertura del Navegador e Intervención Inicial
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
import os

# Conexión manual al monitor virtual
os.environ["DISPLAY"] = ":99"

# 9 # --- CONFIGURACIÓN DEL ENTORNO VISUAL ---
options = Options()
options.binary_location = "/usr/bin/brave-browser"

# --- ARGUMENTOS DE COMPATIBILIDAD CRÍTICA ---
options.add_argument("--no-sandbox")
options.add_argument("--disable-setuid-sandbox") # <--- Línea clave para permisos
options.add_argument("--disable-dev-shm-usage")
options.add_argument("--no-zygote")              # <--- Evita que el proceso muera al nacer
options.add_argument("--disable-gpu")
options.add_argument("--remote-debugging-port=9222")

# Mantenemos el User-Agent de la guía
options.add_argument("user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36")

# 16 # Iniciar el driver
try:
    print("🔋 Iniciando motor de auditoría energética...")
    driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)
    
    # 19 # Navegar al Portal de Energía
    print("🌍 Abriendo portal del Coordinador Eléctrico (Datos Abiertos)...")
    driver.get("https://www.coordinador.cl/sistemas-de-informacion/datos-abiertos/")

    # 23 # --- PAUSA DE INTERVENCIÓN HUMANA ---
    print("\n⚡ ACCIÓN REQUERIDA:")
    print("1. Ve a la pestaña del escritorio del contenedor (noVNC en puerto 6080).")
    print("2. Si la página cargó, interactúa o acepta términos si es necesario.")
    
    input("\n⌨️ Presiona ENTER en esta celda cuando estés listo para continuar...")

    # 31 # --- VERIFICACIÓN DE BLOQUEO (Capítulo 7.4) ---
    if "datos" in driver.page_source.lower():
        print("✅ Confirmado: Portal de energía listo y datos detectados.")
    else:
        print("⚠️ Advertencia: La página cargó pero no se detectan tablas de datos.")

except Exception as e:
    print(f"❌ El error persiste: {e}")
    
```
