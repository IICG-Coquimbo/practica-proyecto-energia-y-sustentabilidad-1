```python
# --- SCRIPT DE CIERRE: TICKET DE SALIDA ---
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, avg, count

# 1. Conexión a la base de datos
spark = SparkSession.builder.appName("TicketSalida_Thalia").getOrCreate()

# Simulamos la carga de los datos procesados en la auditoría
# (Si lograste guardar en Mongo, usa .load(), si no, usamos estos datos de prueba para la entrega)
data = [
    ("Solar PV", 450.5), ("Solar PV", 320.0), ("Solar Concentrada", 120.4),
    ("Eolica", 210.8), ("Eolica", 195.2), ("Hidraulica", 500.1)
]
df = spark.createDataFrame(data, ["tecnologia", "valor"])

# 2. Transformación y Agregación (Precio promedio por "Marca" / Tecnología)
reporte_final = df.groupBy("tecnologia").agg(
    avg("valor").alias("promedio_generacion"),
    count("valor").alias("total_registros")
).orderBy(col("promedio_generacion").desc())

# 3. Resultado para la entrega
reporte_final.show()

# TOTAL DE PRODUCTOS PROCESADOS: 6 registros de matriz energética.
```
